```python
emb_dic = {}
with open('./glove.6B/glove.6B.100d.txt',"r",encoding='utf-8') as f:
    for item in f:
        word_em = item.split()
        word = word_em[0]
        em = []
        for i in range(1,len(word_em)):
            em.append(float(word_em[i]))
        emb_dic[word] = em
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('paraphrase-distilroberta-base-v2')
```

```python
import pandas as pd
import numpy as np
import random
from ast import literal_eval
from scipy import spatial
import re
df = pd.read_csv('ELCo.csv')
df['emoji_list'] = df['emoji_list'].apply(literal_eval)
```

```python
def get_cos_similar_matrix(v1, v2):
    return 1 - spatial.distance.cosine(v1, v2)

def get_similarity(words, emoji, method):
    emoji_matrix = []
    word_matrix = []
    if method == 'gloVe':
        word_list = words.split()
        for word in word_list:
            if word.lower() not in emb_dic.keys():
                print(f"donot have word - {word}")
                continue
            word_matrix.append(emb_dic[word.lower()])
        word_embeddings = np.mean(word_matrix, axis=0)
        for emoji_ in emoji:
            emoji_sep = re.split('_|-',emoji_[1:-1])
            for emoji__ in emoji_sep:
                if emoji__.lower() not in emb_dic.keys():
                    print(f"donot have word - {emoji__} in emoji - {emoji_}")
                    continue
                emoji_matrix.append(emb_dic[emoji__.lower()])
        emoji_embeddings = np.mean(emoji_matrix, axis=0)
    elif method == 'sentence_transformer':
        word_embeddings = word_encode[words]
        emoji_sentence = ''
        for emoji_ in emoji:
            emoji_sep = re.split('_|-',emoji_[1:-1])
            emoji_sentence_ = ' '.join(emoji_sep)
            emoji_sentence += emoji_sentence_ + ' '
        emoji_embeddings = emoji_encode[emoji_sentence]
    return get_cos_similar_matrix(word_embeddings, emoji_embeddings)

def get_most_relevant(word_list, emoji, method, k):
    similarity_dict = {}
    for word in word_list:
        similarity_dict[word] = get_similarity(word, emoji, method)
    result_dic = {}
    result_dic.update(dict(sorted(similarity_dict.items(), key=lambda x: x[1], reverse=True)[:k]))
    return result_dic
```

```python
word_list_all = df['English'].unique()
emoji_list_all = df['emoji_list']
word_encode = {}
emoji_encode = {}
for word in word_list_all:
    word_encode[word] = model.encode(word)
for emoji in emoji_list_all:
    emoji_sentence = ''
    for emoji_ in emoji:
        emoji_sep = re.split('_|-',emoji_[1:-1])
        emoji_sentence_ = ' '.join(emoji_sep)
        emoji_sentence += emoji_sentence_ + ' '
    emoji_encode[emoji_sentence] = model.encode(emoji_sentence)
```
